```python
#dropping player name and adding total matches feature
df = df.drop('Player', axis = 1)
df['total'] = df['Wins'] + df['Losses']
```

```python
df.corr().Winnings
```

```python
#Let's see what feature corresponds the most with winnings in a linear regression model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
best_single_regression = {}
for col in df.columns:
    X = df[[col]]
    y = df[['Winnings']]
    x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 1)
    mlr = LinearRegression()
    mlr.fit(x_train, y_train)
    y_predicted = mlr.predict(x_test)
    best_single_regression.update({'Winnings, Feature: {col}'.format(col=col):mlr.score(x_test, y_test)})
```

```python
df_test = pd.DataFrame.from_dict(best_single_regression, orient='index', columns = ['score'])
print(df_test.sort_values('score'))
```

```python
#Now a multiple regression with total_matches and wins
X = df[['total', 'Wins']]
y = df[['Winnings']]
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 1)
mlr = LinearRegression()
mlr.fit(x_train, y_train)
y_predicted = mlr.predict(x_test)
print(mlr.score(x_test, y_test))
```

```python
#Now using all available features
from sklearn.model_selection import train_test_split
X = df.drop('Winnings', axis=1)
y = df[['Winnings']]
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 1)
mlr = LinearRegression()
mlr.fit(x_train, y_train)
y_predicted = mlr.predict(x_test)
print(mlr.score(x_test, y_test))
```

```python
#Let's see if dropping some of the lower performing features helps the score
X = df.drop(['Winnings', 'Year', 'FirstServe', 'BreakPointsConverted'] , axis=1)
y = df[['Winnings']]
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 1)
mlr = LinearRegression()
mlr.fit(x_train, y_train)
y_predicted = mlr.predict(x_test)
print(mlr.score(x_test, y_test))

#Interestingly this seems to hurt the score and worse than would be predicted based on the individual features dropped.
```
